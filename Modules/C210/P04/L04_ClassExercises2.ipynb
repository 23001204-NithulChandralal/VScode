```python
import pandas as pd 
df = pd.read_excel("gdpData.xlsx")

df.info()
```

```python
# Exercise #1a - Reshaping Data (melt())

df_m =pd.melt(df,id_vars="Country Name",
              var_name="Year",
              value_name="GDP")

df_m
```

```python
# Exercise #1b - Reshaping Data (pivot() & columns)

df_p = df_m.pivot(index="Country Name",
              columns="Year",
              values="GDP")

df_p
```

```python
df_p.drop(columns=[1990,2000], inplace=True)

df_p
```

```python
df_p.sort_values([2013])
```

```python
df_p.sort_values([2021],ascending=False)
```

```python
# Exercise #2a - Grouping Data
mean_year = df_m.groupby('Year')['GDP'].mean()
max = mean_year.max()
print(mean_year)
print("The highest GDP average by year:",max)
mean_country = df_m.groupby("Country Name")["GDP"].mean()
min_country = mean_country.min()
print(mean_country)
print("The lowest GDP average by country:",min_country)
```

```python
# Exercise #2b - Binning Data

#bin the GDP growth data
bins = [0, 5, 10, 15]
GDP = pd.cut(df_m['GDP'], bins)
value_counts_series = GDP.value_counts().rename('GDPGrowth%')
print(value_counts_series)
```

```python
# Exercise #3 - Plotting Data

df_msia = df_m.loc[df_m["Country Name"] == "Malaysia"]
df_msia.plot(y="GDP",x="Year",kind="bar",title="GDP Growth for Malaysia", ylabel="GDP Growth", xlabel="Year")
```
