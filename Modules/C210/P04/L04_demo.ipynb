```python
import pandas as pd
```

```python
name_list = [
                ["Jason" , 5400, 8900, 7265],
                ["James" , 3000, 4030, 7500],
                ["Joseph" , 6701, 5401, 6193]
            ]
cols = ["Name", "Jan", "Feb", "Mar"]
 
df = pd.DataFrame(name_list, columns=cols)
 
df
```

```python
df_m = pd.melt(df,id_vars="Name") #identifier variable


df_m
```

```python
df_m = pd.melt(df,id_vars="Name", #identifier variable
               var_name="Month",  #identifier variable
               value_name="Steps") #identifier variable

df_m
```

```python

df_p = df_m.pivot(index="Name", #identifier variable
               columns="Month",  #identifier variable
               values="Steps") #identifier variable

df_p
```

```python
df_p.drop(columns=["Jan"])
```

```python
df_m.groupby("Month")["Steps"].mean()
```

```python
bins = [2000,4000,6000,8000,10000]
steps = pd.cut(df_m["Steps"],bins)
print(steps.value_count())

value_counts_series = steps.value_counts().rename("Steps")
print(value_counts_series)
```
