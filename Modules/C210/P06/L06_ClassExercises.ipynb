```python
# Exercise #3 - Single Inheritance
# 1. Write the Person class that includes the constructor method to initialize all the properties

class Person:
    def __init__(self,nric,name,address):
        self.__nric = nric
        self.__name = name
        self.__address = address

    def __str__(self):
        out_str = "The Nric of " + self.__name + "is " + self.__nric 
        out_str += ", The address : " + self.__address
        return out_str
        
    def getNRIC(self):
        return self.__nric
    
    def getName(self):
        return self.__name
    
    def getAddress(self):
        return self.__address
    
    def setAddress(self, new_address):
        self.__address = new_address
```

```python
# Exercise #3 - Single Inheritance
# 1. Write the Student class that includes the constructor method to initialize all the properties

class Student(Person):
    def __init__(self, nric, name, address, studentID, admitYear):
        super().__init__(nric, name, address)
        self.__studentID = studentID
        self.__admitYear = admitYear

    def __str__(self):
        out_str = "The NRIC of " + self.getName() + "is " + self.getNRIC() + ", The address : " + self.getAddress()
        out_str += "\nAdmitYear is " + str(self.__admitYear)   
        return out_str

    def getStudentID(self):
        return self.__studentID
    
    def getAdmitYear(self):
        return self.__admitYear
    
    def setAdmitYear(self, new_admitYear):
        self.__admitYear = new_admitYear
```

```python
# Exercise #4 - Hierarchical Inheritance
# 1. Write the Teacher class that includes the constructor method to initialize all the properties

class Teacher(Person):
    def __init__(self, nric, name, address, staffID, salary):
        super().__init__(nric, name, address)
        self.__staffID = staffID
        self.__salary = salary

    def __str__(self):
        out_str = "The NRIC of " + self.getName() + "is " + self.getNRIC() + ", The address : " + self.getAddress()
        out_str += "\nStaffID :" + str(self.__staffID)
        out_str += "\nSalary:" + str(self.__salary)
        return out_str

    def getStaffID(self):
        return self.__staffID
    
    def getSalary(self):
        return self.__salary
    
    def setSalary(self, new_salary):
        self.__salary = new_salary
```
