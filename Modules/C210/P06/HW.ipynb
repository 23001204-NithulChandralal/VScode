```python
#Q1

class Stack:
    def __init__(self):
        self.__data = []

    def __str__(self):
        out_str = "Stack: " + str(self.__data) + "\n"
        out_str += "Is stack empty? : " + str(self.isEmpty()) + "\n"
        popped_item = self.pop()
        if popped_item == 0:
            out_str += "No popped items\n"
        else:
            out_str += "Popped item: " + str(popped_item) + "\n"

        return out_str

    def push(self, item):
        self.__data.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.__data.pop()
        else:
            test1 = print("Empty Stack!")
            return test1

    def isEmpty(self):
        return len(self.__data) == 0

    def __len__(self):
        return len(self.__data)


stack = Stack()

print(stack)
stack.push(5)
print(stack)
stack.push(10)
print(stack)
stack.push(15)
print(stack)
```

```python
#Q2

class Shape:
    def __init__(self, colour, filled):
        self.__colour = colour
        self.__filled = filled

    def __str__(self):
        out_str = "Shape Colour= " + str(self.__colour) +  ", Filled= " + str(self.__filled)
        return out_str

    def getColour(self):
        return self.__colour

    def setColour(self, colour):
        self.__colour = colour

    def isFilled(self):
        return self.__filled

    def setFilled(self, filled):
        self.__filled = filled


test1 = Shape("Red","True")
print(test1)  

test1.setColour('blue')
test1.setFilled("False")
print(test1)  
```

```python
#Q3
class Circle(Shape):
    def __init__(self, radius=1.0, colour='red', filled=True):
        super().__init__(colour, filled)
        self.__radius = radius

    def __str__(self):
        shape_info = super().__str__()
        out_str = shape_info + ", Radius: " + str(self.__radius) + ", Area: " + str(self.getArea()) + ", Perimeter: " + str(self.getPerimeter())
        return out_str

    def getRadius(self):
        return self.__radius

    def setRadius(self, radius):
        self.__radius = radius

    def getArea(self):
        area = 3.14159 * self.__radius * self.__radius
        return area

    def getPerimeter(self):
        peri = 2 * 3.14159 * self.__radius
        return peri

#SHAPE
my_shape = Shape("red","True")
print(my_shape)  
my_shape.setColour('blue')
my_shape.setFilled("False")
print(my_shape)  
print("")

#CIRCLE
my_circle = Circle(1.0,"red","True")
print(my_circle) 
my_circle.setRadius(5.0)
print(my_circle)  

```

```python
#Q4

class Rectangle(Shape):
    def __init__(self, width, length, colour, filled):
        super().__init__(colour, filled)
        self.__width = width
        self.__length = length

    def __str__(self):
        shape_info = super().__str__()
        out_str = shape_info + ", Width= " + str(self.__width) + ", Length= " + str(self.__length) + ", Area= " + str(self.getArea()) + ", Perimeter= " + str(self.getPerimeter())
        return out_str
        
    def getWidth(self):
        return self.__width

    def setWidth(self, width):
        self.__width = width

    def getLength(self):
        return self.__length

    def setLength(self, length):
        self.__length = length

    def getArea(self):
        area1 = self.__width * self.__length
        return area1

    def getPerimeter(self):
        peri1 = 2 * (self.__width + self.__length)
        return peri1


#SHAPE
my_shape = Shape("red","True")
print(my_shape)  
my_shape.setColour('blue')
my_shape.setFilled("False")
print(my_shape)  
print("")

#CIRCLE
my_circle = Circle(1.0,"red","True")
print(my_circle) 
my_circle.setRadius(5.0)
print(my_circle)  
print("")

#RECTANGLE
my_rec = Rectangle(1.0,1.0,"Red", "True")
print(my_rec)
my_rec.setLength(5.0)
my_rec.setWidth(5.0)
print(my_rec)
print("")
```
