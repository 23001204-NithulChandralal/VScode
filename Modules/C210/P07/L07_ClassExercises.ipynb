```python
# Class - Passenger with __init__ and __str__ methods

class Passenger:
    def __init__(self, passenger_id, name, gender, ticket_id):
        self.passenger_id = passenger_id
        self.name = name
        self.gender = gender
        self.ticket_id = ticket_id

    def __str__(self):
        out_str = "Passenger ID: " + self.passenger_id
        out_str += "\nName: " + self.name
        out_str += "\nGender: " + self.gender
        out_str += "\nTicket ID: " + str(self.ticket_id)
        return out_str

    def get_passenger_id(self):
        return self.passenger_id

    def set_ticket(self, ticket_id):
        self.ticket_id = ticket_id
```

```python
# Test stub - Passenger
# 1. Create a Passenger

p1 = Passenger('P001', 'John Smith', 'Male','111')
print(p1)
print("")
p1.set_ticket(222)
print(p1)
print("")
passenger_id = p1.get_passenger_id()
print("Passenger ID:", passenger_id)

```

```python
# Class - Flight with __init__ and __str__ methods


class Flight:
    def __init__(self, flight_id, airline, source, destination, date, start_time, reach_time):
        self.flight_id = flight_id
        self.airline = airline
        self.source = source
        self.destination = destination
        self.date = date
        self.start_time = start_time
        self.reach_time = reach_time  

    def __str__(self):
        out_str = str(self.flight_id) + "\t"
        out_str += str(self.airline) + "\t"
        out_str += str(self.source)  + "\t" + "to    "
        out_str += str(self.destination) + "\t"
        out_str += self.date.strftime("%x") + "\t"
        out_str += self.start_time.strftime("%X") + "\t"
        out_str += self.reach_time.strftime("%X") + "\t"
        return out_str

    def get_flight_id(self):
        return self.flight_id
```

```python
# Test stub - Flight
# 1. Create a Flight

f1 = Flight('SQ123','Singapore Airlines','Singapore','Bangkok', dt.date(2023,6,1), dt.time(3,45,42), dt.time(6,30,45))
f1_id = f1.get_flight_id()
print(f1_id)
```

```python
# Class - Ticket with __init__ and __str__ methods

class Ticket:
    def __init__(self, passenger, flight, ticketID, price):
        self.ticketID = ticketID
        self.passenger = passenger
        self.flight = flight
        self.price = price

    def __str__(self):
        out_str = "TicketID: " + str(self.ticketID) + "\n"
        out_str += "\n*****Passenger Details*****\n" + str(self.passenger) + "\n"
        out_str += "\n*****Flight Details*****\n" + str(self.flight)
        out_str += "\nPrice: " + str(self.price)
        return out_str

    def getTicketID(self):
        return self.ticketID
```

```python
# Class - AirportManagement with __init__ and __str__ methods

class AirportManagement:
    def __init__(self, airportName, cityname, country):
        self.airportName = airportName
        self.cityname = cityname
        self.country = country
        self.flights = {}
        self.passengers = {}
        self.tickets = {}

    def displayAirportDetails(self):
        print("Airport Name:", self.airportName)
        print("City:", self.cityname)
        print("Country:", self.country)
        print("Number of Flights:", len(self.flights))
        print("Number of Passengers:", len(self.passengers))
        print("Number of Tickets:", len(self.tickets))

    def addFlights(self, flight):
        self.flights[flight.get_flight_id()] = flight

    def displayFlights(self):
        print("Flights:")
        for flight_id, flight in self.flights.items():
            print(flight)

    def addPassenger(self, passenger):
        self.passengers[passenger.get_passenger_id()] = passenger

    def displayPassengers(self):
        print("Passengers:")
        for passenger_id, passenger in self.passengers.items():
            print(passenger)

    def addTicket(self, ticket):
        self.tickets[ticket.getTicketID()] = ticket

    def displayTickets(self):
        print("Tickets:")
        for ticket_id, ticket in self.tickets.items():
            print(ticket)

    def __str__(self):
        out_str = "Airport Name: " + str(self.airportName) + "\n"
        out_str += "City: " + str(self.cityname) + "\n"
        out_str += "Country: " + str(self.country) + "\n"
        return out_str
```

```python
# Test stub - AirportManagement
# 1. Create an Airport (any airport) & display the details
# 2. Add the above Passenger to the Airport & display
# 3. Add the above Flight to the Airport & display
# 4. Add the above Ticket to the Airport & display

changi = AirportManagement('Changi','Singapore','Singapore')
print("Airport Details:")
print(changi)

p1 = Passenger('P001', 'John Smith', 'Male','111')
f1 = Flight('SQ123','Singapore Airlines','Singapore','Bangkok', dt.date(2023,6,1), dt.time(3,45,42), dt.time(6,30,45))
t1 = Ticket(passenger=p1, flight=f1, ticketID="T001", price=150.0)

# Add Passenger to the Airport and display passengers
changi.addPassenger(p1)
print("Passengers at the Airport:")
changi.displayPassengers()
print()

# Add Flight to the Airport and display flights
changi.addFlights(f1)
print("Flights at the Airport:")
changi.displayFlights()
print()


# Add Ticket to the Airport and display tickets
changi.addTicket(t1)
print("Tickets at the Airport:")
changi.displayTickets()
```

```python
# Test Stub - Put all together

# Create airport
changi = AirportManagement('Changi','Singapore','Singapore')
changi.displayAirportDetails()
print("")

# Create and add flights to airport
f1 = Flight('SQ123','Singapore Airlines','Singapore','Bangkok' \
            , dt.date(2023,6,1), dt.time(3,45,42), dt.time(6,30,45))
f2 = Flight('JAL832','Japan Airlines','Singapore','Tokyo' \
            , dt.date(2023,6,12), dt.time(1,30,00), dt.time(7,45,15))
changi.addFlights(f1)
changi.addFlights(f2)
changi.displayFlights()

# Create passengers and add passengers to airport
p1 = Passenger('P001', 'John Smith', 'Male',"111")
p2 = Passenger('P002', 'John Leong', 'Male',"222")
changi.addPassenger(p1)
changi.addPassenger(p2)
print("")
changi.displayPassengers()

# Sell tickets to passengers for a flight and add ticket to airport
t1 = Ticket('T001', p1, f1, 450.55)
t2 = Ticket('T002', p2, f2, 470)
changi.addTicket(t1)
changi.addTicket(t2)
changi.displayTickets()

print("")
changi.displayAirportDetails()
```
