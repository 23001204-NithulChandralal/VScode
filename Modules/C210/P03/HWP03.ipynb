```python
#Q2
import pandas as pd
df = pd.read_excel("nba.xlsx")
df.info()
pd.set_option("display.max_rows",500)
pd.set_option('display.max_columns', None)

```

```python
#2a) Determine if there is duplicated data. If duplicated data is found, advise on the number of such duplicated rows and remove such duplicated rows.
print(df[df.duplicated()])    #ROW 254 AND 380 is Duplicated
df.drop_duplicates(inplace=True)
print("---------------------------------------------------")

df.info()       
print("---------------------------------------------------")
```

```python
#2b)identify all the null or empty values in the “College” column and replace them with the string “No college”. 

print(df.info())
df["College"].fillna("No college",inplace=True)
print("---------------------------------------------------")
print(df.info())       
print("---------------------------------------------------")

```

```python
"""
2c) Select all the rows that identify up and coming NBA star players. The basic criteria is players whose age is less than or equal to 25.
Show their name, team and age. Save the data to another spreadsheet named “nba_young_stars.xlsx”
"""
young_stars = df[df['Age'] <= 25]
young_stars = young_stars[['Name', 'Team', 'Age']]
print(young_stars)

young_stars.to_excel('nba_young_stars.xlsx', index=False) 
```
