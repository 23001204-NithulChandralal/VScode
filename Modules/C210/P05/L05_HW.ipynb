```python
# Q3

class Membership:
    def __init__(self):
        self.members = {}

    def __str__(self):
        out_str = ""
        if len(self.members) == 0:
            out_str = "There are no members in the country club"
        else:
            out_str = "Members of the country club:\n"
            for member_name in self.members:
                out_str += member_name + "\n"
        return out_str

    def add(self, person):
        self.members[person.name] = person

    def remove(self, name):
        if name in self.members:
            del self.members[name]

    def search(self, name):
        return name in self.members
```
