```python
"""
Q2.
"""
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        return "Riding"


class Boat(Vehicle):
    def move(self):
        return "Sailing"


class Plane(Vehicle):
    def move(self):
        return "Flying"


# Test the classes and their corresponding polymorphic methods
vehicle = Vehicle("Generic", "Vehicle")
boat = Boat("BoatBrand", "BoatModel")
plane = Plane("PlaneBrand", "PlaneModel")

print(vehicle.move())  
print(boat.move())     
print(plane.move())    
```

```python
"""
#Q3.
"""
import pickle

class AirportManagement:
    def __init__(self, airportName, cityname, country):
        self.airportName = airportName
        self.cityname = cityname
        self.country = country
        self.flights = {}
        self.passengers = {}
        self.tickets = {}
        self.employees = {}
        self.flights_file = "flights.txt"
        self.passengers_file = "passengers.txt"
        self.tickets_file = "tickets.txt"
        self.employees_file = "employees.txt"

        self.load()  # Load data on initialization

    def displayAirportDetails(self):
        print("Airport Name:", self.airportName)
        print("City:", self.cityname)
        print("Country:", self.country)
        print("Number of Flights:", len(self.flights))
        print("Number of Passengers:", len(self.passengers))
        print("Number of Tickets:", len(self.tickets))

    def addFlights(self, flight):
        self.flights[flight.get_flight_id()] = flight

    def displayFlights(self):
        print("Flights:")
        for flight_id, flight in self.flights.items():
            print(flight)

    def addPassenger(self, passenger):
        self.passengers[passenger.get_passenger_id()] = passenger

    def displayPassengers(self):
        print("Passengers:")
        for passenger_id, passenger in self.passengers.items():
            print(passenger)

    def addTicket(self, ticket):
        self.tickets[ticket.getTicketID()] = ticket

    def displayTickets(self):
        print("Tickets:")
        for ticket_id, ticket in self.tickets.items():
            print(ticket)

    def addEmployee(self, employee):
        self.employees[employee.get_employee_id()] = employee

    def displayEmployees(self):
        print("Employees:")
        for employee_id, employee in self.employees.items():
            print(employee)

    def save(self, flights_file=None, passengers_file=None, tickets_file=None, employees_file=None):
            if flights_file:
                with open(flights_file, 'wb') as file:
                    pickle.dump(self.flights, file)
                print(flights_file, "saved successfully.")

            if passengers_file:
                with open(passengers_file, 'wb') as file:
                    pickle.dump(self.passengers, file)
                print(passengers_file, "saved successfully.")

            if tickets_file:
                with open(tickets_file, 'wb') as file:
                    pickle.dump(self.tickets, file)
                print(tickets_file, "saved successfully.")

            if employees_file:
                with open(employees_file, 'wb') as file:
                    pickle.dump(self.employees, file)
                print(employees_file, "saved successfully.")

            if not any((flights_file, passengers_file, tickets_file, employees_file)):
                print("No valid filenames provided.")
```
