```python
# Exercise #1 – Polymorphism - Overriding
# 1. Copy Person class from L06
# 2. Add the displayDetails() method

class Person:
    def __init__(self,nric,name,address):
        self.__nric = nric
        self.__name = name
        self.__address = address

    def __str__(self):
        out_str = "The Nric of " + self.__name + "is " + self.__nric 
        out_str += ", The address : " + self.__address
        return out_str
        
    def getNRIC(self):
        return self.__nric
    
    def getName(self):
        return self.__name
    
    def getAddress(self):
        return self.__address
    
    def setAddress(self, new_address):
        self.__address = new_address

    #new method
    def displayDetails(self):           
        print("NRIC:", self.__nric)
        print("Name:", self.__name)
        print("Address:", self.__address)        
```

```python
# Exercise #2a - Storing Objects Persistently
# 1. Copy Dealership class from L05
# 2. Add a new method save()

import pickle

class Dealership:
    def __init__(self):
        self.cars = {}  #create a dictionary to store cars

    def add(self,new_car):
        #to get PlateNo of the new_car
        plateNo = new_car.getPlateNo()
        self.cars[plateNo] = new_car

    def __str__(self):
        out_str =""
        if len(self.cars) == 0:
            out_str = "There is no car under the dealership"
        else:
            out_str = "The cars under the dealership: \n"  #\n is to create new line
            for plateNo in self.cars:
                out_str += plateNo + "\n"
        return out_str
    
    def remove(self, plate_no):
        if plate_no in self.cars:
            del self.cars[plate_no]
        else:
            print("Car with plate number",plate_no,"not found")

    #new method
    def save(self,filename=None):
        if filename is None:
            print("No filename provided.")
            return
        else:
            with open(filename,'wb') as file: # w - write, b - binary format
                for car in self.cars:
                    the_car = self.cars[car]
                    pickle.dump(the_car,file)      #write "the_Car" to the file

            print(filename , "saved succesfully.")
    
        
```

```python
# Exercise #2b – Loading Objects from a File
# 1. Update __init__ of Dealership class

import pickle

class Dealership:
    def __init__(self,filename=None):
        self.cars = {}  #create a dictionary to store cars

        #new codes to read from the file
        if filename != None:
            with open(filename, 'rb') as file: # r - to read 
                while True:
                    try:
                        car = pickle.load(file)
                        self.add(car)   #add to the dictionary 
                    except EOFError: # when End-Of-File reached 
                        break 
            print("Loading from", filename, "completed.")

    def add(self,new_car):
        #to get PlateNo of the new_car
        plateNo = new_car.getPlateNo()
        self.cars[plateNo] = new_car

    def __str__(self):
        out_str =""
        if len(self.cars) == 0:
            out_str = "There is no car under the dealership"
        else:
            out_str = "The cars under the dealership: \n"  #\n is to create new line
            for plateNo in self.cars:
                out_str += plateNo + "\n"
        return out_str
    
    def remove(self, plate_no):
        if plate_no in self.cars:
            del self.cars[plate_no]
        else:
            print("Car with plate number",plate_no,"not found")

    #new method
    def save(self,filename=None):
        if filename is None:
            print("No filename provided.")
            return
        else:
            with open(filename,'wb') as file: # w - write, b - binary format
                for car in self.cars:
                    the_car = self.cars[car]
                    pickle.dump(the_car,file)    #write "the_Car" to the file

            print(filename , "saved succesfully.")
    
```

```python
# Exercise #3 - Enhanced AMS (refer to the class diagram)
# 1.  Write the Employee class that includes the constructor method to initialize all the properties

class Employee:
    def __init__(self,employeeID,name,salary, designation):
        self.__employeeID = employeeID
        self.__name = name 
        self.__salary = salary 
        self.__designation =  designation

    def getEmployeeID(self):
        return self.__employeeID
    
    def __str__(self):
        out_str = self.__employeeID + '\t'
        out_str += self.__name + '\t'
        out_str += str(self.__salary) + '\t'
        out_str += self.__designation 
        return out_str 

```

```python
# Exercise #3 - Enhanced AMS (refer to the class diagram)
# 1.  Write the AirportEmployee class that includes the constructor method to initialize all the properties

class AirportEmployee(Employee):
    def __init__(self,employeeID,name,salary, designation, dept):
        super().__init__(employeeID,name,salary, designation)
        self.__dept = dept 


    def __str__(self):
        return super().__str__() + '\t' + self.__dept 

```

```python
# Exercise #3 - Enhanced AMS (refer to the class diagram)
# 1.  Write the AirlineEmployee class that includes the constructor method to initialize all the properties

class AirlineEmployee(Employee):
    def __init__(self,employeeID,name,salary, designation, airline):
        super().__init__(employeeID,name,salary, designation)
        self.__airline = airline

    def __str__(self):
        return super().__str__() + '\t' + str(self.__airline) 


```

```python
# Exercise #3 - Enhanced AMS
# 1. Copy all classes from L07 except AirportManagement

class Ticket:
    def __init__(self, passenger, flight, ticketID, price):
        self.ticketID = ticketID
        self.passenger = passenger
        self.flight = flight
        self.price = price

    def __str__(self):
        out_str = "TicketID: " + str(self.ticketID) + "\n"
        out_str += "\n*****Passenger Details*****\n" + str(self.passenger) + "\n"
        out_str += "\n*****Flight Details*****\n" + str(self.flight)
        out_str += "\nPrice: " + str(self.price)
        return out_str

    def getTicketID(self):
        return self.ticketID
    
class Flight:
    def __init__(self, flight_id, airline, source, destination, date, start_time, reach_time):
        self.flight_id = flight_id
        self.airline = airline
        self.source = source
        self.destination = destination
        self.date = date
        self.start_time = start_time
        self.reach_time = reach_time  

    def __str__(self):
        out_str = str(self.flight_id) + "\t"
        out_str += str(self.airline) + "\t"
        out_str += str(self.source)  + "\t" + "to    "
        out_str += str(self.destination) + "\t"
        out_str += self.date.strftime("%x") + "\t"
        out_str += self.start_time.strftime("%X") + "\t"
        out_str += self.reach_time.strftime("%X") + "\t"
        return out_str

    def get_flight_id(self):
        return self.flight_id
    
class Passenger:
    def __init__(self, passenger_id, name, gender, ticket_id):
        self.passenger_id = passenger_id
        self.name = name
        self.gender = gender
        self.ticket_id = ticket_id
        
    def __str__(self):
        out_str = "Passenger ID: " + self.passenger_id
        out_str += "\nName: " + self.name
        out_str += "\nGender: " + self.gender
        out_str += "\nTicket ID: " + str(self.ticket_id)
        return out_str

    def get_passenger_id(self):
        return self.passenger_id

    def set_ticket(self, ticket_id):
        self.ticket_id = ticket_id
```

```python
# Exercise #3 - Enhanced AMS (refer to the class diagram)
# 1. Copy AirportManagement class from L07
# 2. Add a new attribute employees{str, employee}
# 3. Add a new new method addEmployee()
# 4. Add a new method displayEmployee()

# Class - AirportManagement with __init__ and __str__ methods

import datetime as dt 

class AirportManagement:
    def __init__(self, airportName, cityname, country):
        self.airportName = airportName
        self.cityname = cityname
        self.country = country
        self.flights = {}
        self.passengers = {}
        self.tickets = {}

        #new attribute 
        self.__employees = {}

    def displayAirportDetails(self):
        print("Airport Name:", self.airportName)
        print("City:", self.cityname)
        print("Country:", self.country)
        print("Number of Flights:", len(self.flights))
        print("Number of Passengers:", len(self.passengers))
        print("Number of Tickets:", len(self.tickets))

    def addFlights(self, flight):
        self.flights[flight.get_flight_id()] = flight

    def displayFlights(self):
        print("Flights:")
        for flight_id, flight in self.flights.items():
            print(flight)

    def addPassenger(self, passenger):
        self.passengers[passenger.get_passenger_id()] = passenger

    def displayPassengers(self):
        print("Passengers:")
        for passenger_id, passenger in self.passengers.items():
            print(passenger)

    def addTicket(self, ticket):
        self.tickets[ticket.getTicketID()] = ticket

    def displayTickets(self):
        print("Tickets:")
        for ticket_id, ticket in self.tickets.items():
            print(ticket)

    # new methods 
    def addEmployee(self, employee):
        self.__employees[employee.getEmployeeID()] = employee

    def displayEmployees(self):
            print("\nEmployee Info")
            print("----------------")
            for i in self.__employees:
                print(self.__employees[i])

    def __str__(self):
        out_str = "Airport Name: " + str(self.airportName) + "\n"
        out_str += "City: " + str(self.cityname) + "\n"
        out_str += "Country: " + str(self.country) + "\n"
        return out_str

```
